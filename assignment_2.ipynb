```python
import pandas as pd 
```

```python
#  Create a DataFrame with the following data:  
```

```python
data = {
      'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
      'Age': [24, 27, 22, 32, 29],
      'Department': ['HR', 'Finance', 'IT', 'Marketing', 'HR'],
      'Salary': [45000, 54000, 50000, 62000, 47000]
     }
df=pd.DataFrame(data)
df
```

```python
 # a. Print the first five rows of the DataFrame.  
 #  b. Get the summary statistics of the 'Age' and 'Salary' columns.  
 #  c. Calculate the average salary of employees in the 'HR' depart
```

```python
first_5=df.head(5)
first_5
```

```python
summary=df.describe()
summary
```

```python
avg_hr_salary = df[df['Department'] == 'HR']['Salary'].mean()
print("Average HR Salary:", avg_hr_salary)
```

```python
# 2. Add a new column, 'Bonus', which is 10% of the salary. 
```

```python
df['Bonus'] = df['Salary'] * 0.10
print(df)
```

```python
# 3. Filter the DataFrame to show employees aged between 25 and 30.  
```

```python
filtered_df = df[(df['Age'] >= 25) & (df['Age'] <= 30)]
print(filtered_df)
```

```python
# 4. Group the data by 'Department' and calculate the average salary for each department.
```

```python
dept_avg_salary = df.groupby('Department')['Salary'].mean()
print(dept_avg_salary)
```

```python
# 5. Sort the DataFrame by 'Salary' in ascending order and save the result to a new CSV file.  
```

```python
sorted_df = df.sort_values(by='Salary', ascending=True)
# Save to CSV
sorted_df.to_csv('sorted_salary.csv', index=False)
print(sorted_df)
```

```python
import matplotlib.pyplot as plt
import random
```

```python
# 1. Create a simple line plot for the following data:
  
#   x = [1, 2, 3, 4, 5]
#   y = [10, 15, 25, 30, 50]
  
#   a. Plot the data.  
#   b. Customize the plot by adding a title, axis labels, and a grid.  
```

```python
x = [1, 2, 3, 4, 5]
y = [10, 15, 25, 30, 50]
# Plot
plt.figure()
plt.plot(x, y, marker='o')
# Customization
plt.title("Simple Line Plot")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.grid(True)
plt.show()
```

```python
# 2. Create a bar graph to represent the marks scored by students in a subject:  
  
#   students = ['John', 'Jane', 'Alice', 'Bob']
#   marks = [75, 85, 60, 90]
  
#   a. Plot the data as a bar graph.  
#   b. Customize the colors and add a title. 
```

```python
students = ['John', 'Jane', 'Alice', 'Bob']
marks = [75, 85, 60, 90]
# Plot
plt.figure()
plt.bar(students, marks, color=['blue', 'violet', 'orange', 'red'])
# Customization
plt.title("Marks Scored by Students")
plt.xlabel("Students")
plt.ylabel("Marks")

plt.show()
```

```python
# 3. Create a pie chart to represent the percentage distribution of a companys revenue from different regions:  
  
#   regions = ['North America', 'Europe', 'Asia', 'Others']
#   revenue = [45, 25, 20, 10]
  
#   a. Create a pie chart with the region names as labels.  
#   b. Highlight the region with the highest revenue.
```

```python
regions = ['North America', 'Europe', 'Asia', 'Others']
revenue = [45, 25, 20, 10]
explode = [0.1, 0, 0, 0]

# Plot
plt.figure()
plt.pie(
    revenue,
    labels=regions,
    explode=explode,
    autopct='%1.1f%%',
    startangle=90
)

plt.title("Revenue Distribution by Region")

plt.show()
```

```python
# 4. Generate a histogram to show the frequency distribution of randomly generated integers between 1 and 100 (sample size = 1000).
```

```python
data = [random.randint(1, 100) for _ in range(1000)]

# Plot
plt.figure()
plt.hist(data, bins=20)

# Customization
plt.title("Histogram of Random Integers (1100)")
plt.xlabel("Value")
plt.ylabel("Frequency")

plt.show()
```
